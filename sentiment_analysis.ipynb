```python
import numpy as np
import pandas as pd
```

```python
dt=pd.read_csv("consumer_reviews.csv")
dt.head(10)
```

```python
dt["reviews_text"]= dt["reviews.title"]+ " "+ dt["reviews.text"]
```

```python
dt.drop(dt.columns.difference(['reviews.rating','reviews_text']), 1, inplace=True)
```

```python
dt = dt.dropna()
```

```python
conditions = [(dt['reviews.rating']>3), (dt['reviews.rating']==3), (dt['reviews.rating']<3)]
choices = ['positive', 'neutral', 'negative']
dt['sentiment_label']= np.select(conditions, choices, default='neutral')
```

```python
dt['reviews_text']=dt['reviews_text'].apply(remove_special_characters)
dt.head()
```

```python
def tokenizer(text):
    return text.split()
```

```python
dt['reviews_text']=dt['reviews_text'].apply(tokenizer)
dt.head()
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
porter = SnowballStemmer("english", ignore_stopwords=False)
def stemmer(text):
    return [porter.stem(word) for word in text]
```

```python
dt['reviews_text']=dt['reviews_text'].apply(stemmer)
dt['reviews_text'][2]
```

```python
from nltk.stem import WordNetLemmatizer 
```

```python
lemmatizer = WordNetLemmatizer()
def lemmitize_it(text):
    return [lemmatizer.lemmatize(word, pos = "a") for word in text]
```

```python
dt['reviews_text']=dt['reviews_text'].apply(lemmitize_it)
dt['reviews_text'][2]
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
def stop_it(text):
    review = [word for word in text if not word in stop_words ] 
    return review
```

```python
dt['reviews_text']=dt['reviews_text'].apply(stop_it)
dt['reviews_text'][2]
```

```python
dt.head()
```

```python
dt['reviews_text'] = dt['reviews_text'].apply(lambda x: ' '.join(map(str, x)))
dt.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
tfidf = TfidfVectorizer()
y = dt['sentiment_label']
x = tfidf.fit_transform(dt['reviews_text'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.2,shuffle=False)
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)

from sklearn.metrics import accuracy_score
acc_log = accuracy_score(y_pred, y_test)*100
print("accuracy:",acc_log )
```
